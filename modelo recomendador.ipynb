```python
import pandas as pd
import numpy as np
import json
```

```python
with open('review.json', 'r') as f:
    rev= f.read()
```

```python
reviews= json.loads(rev)
```

```python
reviews
```

```python
rev = pd.DataFrame(reviews)
```

```python
rev.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



# Reemplazar los valores faltantes con una cadena vacía
col1 = rev['Nombre encontrado'].fillna('')
col2 = rev['Producto'].fillna('')

# Crear un objeto TfidfVectorizer
tfidf = TfidfVectorizer(token_pattern=r'\b\w+\b')

# Transformar las columnas en vectores TF-IDF
vec1 = tfidf.fit_transform(col1)
vec2 = tfidf.transform(col2)

# Calcular la similitud coseno entre los vectores
similarity = cosine_similarity(vec1, vec2)

# Convertir la matriz de similitud a un DataFrame
similarity_df = pd.DataFrame(similarity, columns=col2.index, index=col1.index)

# Obtener los índices de las filas que tienen similitud mayor a 0.5
similar_rows = similarity_df[similarity_df > 0.5].stack().index.tolist()

# Crear un nuevo DataFrame con las filas seleccionadas
new_data = pd.DataFrame(columns=rev.columns)
for row1, row2 in similar_rows:
    new_data = new_data.append(rev.loc[[row1, row2]])

# Eliminar duplicados y ordenar por índice
new_data = new_data.drop_duplicates().sort_index()

# Imprimir el nuevo DataFrame
print(new_data)

```
