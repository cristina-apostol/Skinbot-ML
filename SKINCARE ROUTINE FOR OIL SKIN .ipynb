```python
import pandas as pd
import numpy as np
import json
import warnings
warnings.filterwarnings("ignore")
import json

from scipy.spatial.distance import pdist, squareform
```

Para realizar un modelo de recomendación vamos a filtrar cada tipo de piel, según tipo de producto, para poder crear una buena rutina de skincare.

## OIL SKIN - SERUM

```python
filtro= skincare.loc[(skincare.Skin == 'oil') & (skincare.Type == 'Serum')]
```

```python
filtro.Products.unique()
```

```python
oil_serum=filtro.set_index('Products')
```

```python
oil_serum
```

```python
oil_serum = oil_serum.drop(columns= ['Type', 'Skin'], axis=1)
```

Creamos un usuario random, para la recomendación de productos con piel GRASA.

```python
random={"Naturium Skincare - Vitamin C Super Serum Plus": 4.5, 
     'Dr. Dennis Gross Skincare - Advanced Retinol + Ferulic Texture Renewal Serum':3,
       'PURITO - Pure Vitamin C Serum':1,
       'Olay - Tone Perfection Serum with Vitamin B3 and Vitamin C':3,
       'The INKEY List - Collagen Booster':3,
       'Urban Skin - HydraFirm + Brightening Serum':2.5,
       'Acure - Brightening Tranexamic Acid Serum':2,
       'Round Lab - Soybean Nourishing Serum':3}
```

```python
oil_serum['random']=pd.Series(random)
```

```python
prueba_serum= oil_serum.loc[oil_serum["random"].isnull()] 
```

```python
rat_serum= prueba_serum.drop('random', axis=1)
```

```python
rating_serum= oil_serum.loc[oil_serum["random"].notnull()]
```

```python
distancias=squareform(pdist(rating_serum.T, 'euclidean')) 

similitud=1/(1+distancias)

similitudes_serum=pd.DataFrame(similitud, index=rating_serum.columns, columns=rating_serum.columns)

similitudes_serum
```

```python
simil_serum=similitudes_serum['random'].sort_values(ascending=False)[1:]

simil_serum
```

```python
for n,s in dict(simil_serum).items():
    
    rat_serum[n]=rat_serum[n]*s   # producto entre similitud y rating
    
rat_serum['Total']=rat_serum.sum(axis=1)

reco_serum=rat_serum.sort_values('Total', ascending=False)
```

```python
reco_serum[:1]
```
