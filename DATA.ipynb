```python
import pandas as pd
import numpy as np
import json
import warnings
warnings.filterwarnings("ignore")
```

```python

with open('dry.json', 'r') as f:
    data= f.read()
with open('oil.json', 'r') as f:
    oil= f.read()
with open('acne.json', 'r') as f:
    acne= f.read()
with open('age.json', 'r') as f:
    age= f.read()
with open('bright.json', 'r') as f:
    bright= f.read()
with open('pores.json', 'r') as f:
    pores= f.read()
with open('scar.json', 'r') as f:
    scar= f.read()
with open('sens.json', 'r') as f:
    sens= f.read()
with open('spots.json', 'r') as f:
    spots= f.read()
with open('text.json', 'r') as f:
    text= f.read()
    
```

```python
dry = json.loads(data)
oil = json.loads(oil)
acne = json.loads(acne)
age = json.loads(age)
bright = json.loads(bright)
pores = json.loads(pores)
scar = json.loads(scar)
sens = json.loads(sens)
spots = json.loads(spots)
text = json.loads(text)
```

```python

oil_keys = oil.keys()
lista_1 = []
lista_2 = []
lista_3 = []

for key in oil_keys:
    if len(oil[key]) > 0:
        lista_1.append(oil[key][0][0][0])
    if len(oil[key]) > 1:
        lista_2.append(oil[key][1][0][0])
    if len(oil[key]) > 2:
        lista_3.append(oil[key][2][0][0])


len(lista_1)
```

```python
data = {'lista_1': lista_1, 'lista_2': lista_2, 'lista_3': lista_3}
```
